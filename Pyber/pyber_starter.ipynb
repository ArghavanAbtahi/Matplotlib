```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File to Load
city_data_to_load = ("data/city_data.csv")
ride_data_to_load = ("data/ride_data.csv")

#Read the City and Ride Data
city_data_read= pd.read_csv(city_data_to_load)
ride_data_read= pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
cityandride = city_data_read.merge(ride_data_read, on="city")
                                
# Display the data table for preview
cityandride.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban = cityandride.loc[(cityandride["type"] == "Urban")]
suburban = cityandride.loc[(cityandride["type"] == "Suburban")]
rural = cityandride.loc[(cityandride["type"] == "Rural")]

averagefareurban = urban.groupby(["city"])["fare"].mean()
averagefaresuburban = suburban.groupby(["city"])["fare"].mean()
averagefarerural = rural.groupby(["city"])["fare"].mean()

numofridesurban = urban.groupby(["city"])["ride_id"].count()
numofridessuburban = suburban.groupby(["city"])["ride_id"].count()
numofridesrural = rural.groupby(["city"])["ride_id"].count()

drivercounturban = urban.groupby(["city"])["driver_count"].unique().str.get(0)
drivercountsuburban = suburban.groupby(["city"])["driver_count"].unique().str.get(0)
drivercountrural = rural.groupby(["city"])["driver_count"].unique().str.get(0)

# Build the scatter plots for each city types

plt.scatter(numofridesurban, averagefareurban, marker="o", facecolors="coral", edgecolors="black",
            s=drivercounturban*10, alpha=.75, label="Urban")
plt.scatter(numofridessuburban, averagefaresuburban, marker="o", facecolors="lightskyblue", edgecolors="black",
            s=drivercountsuburban*10, alpha=0.75, label="Subrban")
plt.scatter(numofridesrural, averagefarerural, marker="o", facecolors="gold", edgecolors="black",
            s=drivercountrural*10, alpha=0.75, label="Rural")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fair ($)")

# Create a legend

plt.legend(title="City Type", loc='best')
plt.grid()

# Incorporate a text label regarding circle size

plt.text(0.93, .6, "note:\n Circle size correlates with driver count per city", 
         fontsize=12,transform=plt.gcf().transFigure)

# Save Figure
plt.savefig("Pyber_Plott")

# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urbanFare = urban["fare"].sum()
suburbanFare = suburban["fare"].sum()
ruralFare = rural["fare"].sum()

labels = ["Urban", "Suburban", "Rural"]
sizes = [urbanFare, suburbanFare, ruralFare]
colors = ["coral", "lightskyblue", "gold"]
explode = (0.1, 0, 0)

# Build Pie Chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Fares by City Type\n")
plt.axis("equal")

# Save Figure
plt.savefig("Fare_By_City.png")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urbanRides = urban["ride_id"].count()
suburbanRides = suburban["ride_id"].count()
ruralRides = rural["ride_id"].count()

labels = ["Urban", "Suburban", "Rural"]
sizes = [urbanRides, suburbanRides, ruralRides]
colors = ["coral", "lightskyblue", "gold"]
explode = (0.1, 0, 0)

# Build Pie Chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Rides by City Type\n")
plt.axis("equal")

# Save Figure
plt.savefig("Rides_By_City.png")
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
urbanDrivers = drivercounturban.sum()
suburbanDrivers = drivercountsuburban.sum()
ruralDrivers = drivercountrural.sum()

labels = ["Urban", "Suburban", "Rural"]
sizes = [urbanDrivers, suburbanDrivers, ruralDrivers]
colors = ["coral", "lightskyblue", "gold"]
explode = (0.1, 0, 0)

# Build Pie Chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Drivers by City Type\n")
plt.axis("equal")

# Save Figure
plt.savefig("Drivers_By_City.png")
plt.show()
```

** Conculsions:
    * the lesser the number of rides per city, the higher the fare.  Rural cities had higher fares.
    * even though the percent of the number of drivers in urban areas is higher, the percent of total rides and 
    total fares is less by 20%
    * since the number of rides and total fares in rural areas is higher than the total drivers in rural areas,
    it could mean that people are coming from suburban and urban areas to rural areas.
